## 3.1 Accessing an Assay Dataset

```python
# Define the URL to download the dataset

url = "https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE81934"
```

```python
# Download the data
local_filename = "GSE81934_RAW.tar"
directory = "./data/"


import requests
import shutil
import os
import numpy as np

local_path = os.path.join(directory, local_filename)

if not os.path.exists(directory):
    os.makedirs(directory)
    print(f"Created directory: {directory}")
else:
    print(f"Directory already exists: {directory}")

# Check if file already exists
if not os.path.isfile(local_path):
    with requests.get(url, stream=True) as r:
        with open(os.path.join(directory, local_filename), 'wb') as f:
            shutil.copyfileobj(r.raw, f)
    print("Download completed.")
else:
    print(f"File ({local_path}) already exists. Skipping download.")

# Extract the tar file
import tarfile

#Check if extracted files already exist (expected file format: *.gz)
extracted_files = [f for f in os.listdir(directory) if f.endswith(".gz")] # list with names of extracted files
if not extracted_files:
    with tarfile.open(local_path, "r") as tar:
        tar.extractall(path=directory)
    print("Extraction completed.")
    print(f"Number of files: {len(extracted_files)}")
else:
    print("Files already exist. Skipping extraction.")
    print(f"Number of files: {len(extracted_files)}")
```

## 3.2 Control the Downloads

```python
# Check the Unique file types
file_endings = set()
for filename in extracted_files:
    ending = filename.split('.')[-2]
    file_endings.add(ending)
print(f"Unique file types: {file_endings}")
```
